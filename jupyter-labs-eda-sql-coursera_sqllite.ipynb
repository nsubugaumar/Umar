### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
import pandas as pd
import sqlite3

con = sqlite3.connect('spacex.db')

# Task 3: Handle different NASA/CRS naming patterns
sql_task3_alt = '''
SELECT 
    SUM("PayloadMass") as "TotalPayloadMass",
    COUNT(*) as "MissionCount"
FROM SPACEXTBL 
WHERE UPPER("Customer") LIKE '%NASA%' 
   OR UPPER("Customer") LIKE '%CRS%'
   OR UPPER("Customer") LIKE '%COMMERCIAL RESUPPLY SERVICES%';
'''

result = pd.read_sql_query(sql_task3_alt, con)

print("ðŸŒŒ NASA/CRS PAYLOAD MASS ANALYSIS")
print("-" * 40)

total_mass = result['TotalPayloadMass'].iloc[0]
mission_count = result['MissionCount'].iloc[0]

if mission_count > 0:
    print(f"\nâœ… Results:")
    print(f"   Missions for NASA/CRS: {mission_count}")
    print(f"   Total payload mass: {total_mass:,.2f} kg")
    print(f"   Average per mission: {total_mass/mission_count:,.2f} kg")
else:
    print("âŒ No NASA/CRS missions found")

con.close()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
import pandas as pd
import sqlite3

# Connect to database
con = sqlite3.connect('spacex.db')

print("=" * 70)
print("TASK 4: Average payload mass carried by booster version F9 v1.1")
print("=" * 70)

# SQL query for Task 4
sql_task4 = '''
SELECT 
    AVG("PayloadMass") as "AveragePayloadMass",
    COUNT(*) as "NumberOfLaunches",
    MIN("PayloadMass") as "MinPayloadMass",
    MAX("PayloadMass") as "MaxPayloadMass"
FROM SPACEXTBL 
WHERE "BoosterVersion" = 'F9 v1.1';
'''

# Execute query
df_result = pd.read_sql_query(sql_task4, con)

# Extract values
avg_mass = df_result['AveragePayloadMass'].iloc[0]
num_launches = df_result['NumberOfLaunches'].iloc[0]
min_mass = df_result['MinPayloadMass'].iloc[0]
max_mass = df_result['MaxPayloadMass'].iloc[0]

# Display result
if num_launches > 0:
    print(f"\nâœ… Results for F9 v1.1:")
    print(f"   â€¢ Number of launches: {num_launches}")
    print(f"   â€¢ Average payload mass: {avg_mass:,.2f} kg")
    print(f"   â€¢ Minimum payload mass: {min_mass:,.2f} kg")
    print(f"   â€¢ Maximum payload mass: {max_mass:,.2f} kg")
    print(f"   â€¢ Range: {max_mass - min_mass:,.2f} kg")
else:
    print(f"\nâŒ No launches found for booster version F9 v1.1")
    print("   Checking available booster versions...")
    
    # Check what booster versions exist
    versions_query = 'SELECT DISTINCT "BoosterVersion" FROM SPACEXTBL ORDER BY "BoosterVersion";'
    versions = pd.read_sql_query(versions_query, con)
    print("\nAvailable booster versions:")
    print(versions.to_string(index=False))

con.close()
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
import pandas as pd
import sqlite3

con = sqlite3.connect('spacex.db')

# Most concise solution assuming exact column names
result = pd.read_sql_query(
    '''SELECT MIN("Date") as FirstDate 
       FROM SPACEXTBL 
       WHERE "Outcome" = 'Success' AND "LandingPad" LIKE '%ground%';''', 
    con
)

print(f"First successful ground landing: {result['FirstDate'].iloc[0]}")
con.close()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
import pandas as pd
import sqlite3

# Connect to database
con = sqlite3.connect('spacex.db')

print("=" * 80)
print("TASK 6: Boosters with drone ship success and payload mass 4000-6000 kg")
print("=" * 80)

# SQL query for Task 6
sql_task6 = '''
SELECT DISTINCT "BoosterVersion"
FROM SPACEXTBL 
WHERE ("Outcome" = 'Success' OR "LandingOutcome" = 'Success')
  AND ("LandingPad" LIKE '%drone%' OR "LandingPad" LIKE '%ASDS%' OR "LandingPad" LIKE '%ship%')
  AND "PayloadMass" > 4000 
  AND "PayloadMass" < 6000
ORDER BY "BoosterVersion";
'''

# Execute query
df_result = pd.read_sql_query(sql_task6, con)

# Display result
if not df_result.empty:
    print(f"\nâœ… Found {len(df_result)} unique booster versions:")
    print("-" * 40)
    
    for i, row in df_result.iterrows():
        print(f"{i+1}. {row['BoosterVersion']}")
    
    print("-" * 40)
else:
    print("\nâŒ No boosters found matching all criteria")
    
    # Let's debug by checking each condition separately
    print("\nðŸ” Debugging - Checking individual conditions:")
    
    # Check 1: Successful missions
    query1 = 'SELECT COUNT(DISTINCT "BoosterVersion") FROM SPACEXTBL WHERE "Outcome" = "Success";'
    result1 = pd.read_sql_query(query1, con)
    print(f"1. Boosters with success: {result1.iloc[0,0]}")
    
    # Check 2: Drone ship landings
    query2 = '''
    SELECT COUNT(DISTINCT "BoosterVersion") 
    FROM SPACEXTBL 
    WHERE "LandingPad" LIKE '%drone%' OR "LandingPad" LIKE '%ASDS%' OR "LandingPad" LIKE '%ship%';
    '''
    result2 = pd.read_sql_query(query2, con)
    print(f"2. Boosters with drone ship: {result2.iloc[0,0]}")
    
    # Check 3: Payload mass range
    query3 = '''
    SELECT COUNT(DISTINCT "BoosterVersion") 
    FROM SPACEXTBL 
    WHERE "PayloadMass" > 4000 AND "PayloadMass" < 6000;
    '''
    result3 = pd.read_sql_query(query3, con)
    print(f"3. Boosters with payload 4000-6000 kg: {result3.iloc[0,0]}")

con.close()
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
import pandas as pd
import sqlite3

con = sqlite3.connect('spacex.db')

# One-liner to get boosters with maximum payload
result = pd.read_sql_query(
    'SELECT DISTINCT "BoosterVersion" FROM SPACEXTBL WHERE "PayloadMass" = (SELECT MAX("PayloadMass") FROM SPACEXTBL);', 
    con
)

print("Boosters with maximum payload:")
print(result.to_string(index=False))

con.close()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
import pandas as pd
import sqlite3

# Connect to database
con = sqlite3.connect('spacex.db')

print("=" * 90)
print("TASK 9: 2015 Drone Ship Failure Records with Month Names")
print("=" * 90)

# SQL query for Task 9
sql_task9 = '''
SELECT 
    substr(Date, 6, 2) as MonthNum,
    CASE substr(Date, 6, 2)
        WHEN '01' THEN 'January'
        WHEN '02' THEN 'February'
        WHEN '03' THEN 'March'
        WHEN '04' THEN 'April'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'June'
        WHEN '07' THEN 'July'
        WHEN '08' THEN 'August'
        WHEN '09' THEN 'September'
        WHEN '10' THEN 'October'
        WHEN '11' THEN 'November'
        WHEN '12' THEN 'December'
        ELSE 'Unknown'
    END as MonthName,
    "BoosterVersion",
    "LaunchSite",
    "LandingOutcome",
    "Outcome",
    "Date",
    "FlightNumber",
    "LandingPad"
FROM SPACEXTBL 
WHERE substr(Date, 0, 5) = '2015'
  AND ("LandingOutcome" LIKE '%Failure%' OR "Outcome" LIKE '%Failure%')
  AND ("LandingPad" LIKE '%drone%' OR "LandingPad" LIKE '%ASDS%' OR "LandingPad" LIKE '%ship%')
ORDER BY MonthNum, "Date";
'''

# Execute query
df_result = pd.read_sql_query(sql_task9, con)

# Display result
if not df_result.empty:
    print(f"\nðŸ“… Found {len(df_result)} drone ship failure records in 2015:")
    print("-" * 90)
    
    # Group by month for better display
    for month_num in sorted(df_result['MonthNum'].unique()):
        month_data = df_result[df_result['MonthNum'] == month_num]
        month_name = month_data['MonthName'].iloc[0]
        
        print(f"\nðŸ“… {month_name} 2015:")
        print("-" * 40)
        
        for _, row in month_data.iterrows():
            print(f"  â€¢ Flight {row['FlightNumber']}")
            print(f"    Booster: {row['BoosterVersion']}")
            print(f"    Launch Site: {row['LaunchSite']}")
            print(f"    Date: {row['Date']}")
            print(f"    Landing Pad: {row['LandingPad']}")
            print(f"    Outcome: {row['Outcome']}")
            print(f"    Landing Outcome: {row['LandingOutcome']}")
            print()
    
    # Summary statistics
    print("\nðŸ“Š SUMMARY STATISTICS:")
    print("-" * 40)
    
    # Count by month
    monthly_counts = df_result.groupby('MonthName').size()
    for month, count in monthly_counts.items():
        print(f"  {month}: {count} failure(s)")
    
    # Count by booster version
    booster_counts = df_result.groupby('BoosterVersion').size()
    print(f"\n  By Booster Version:")
    for booster, count in booster_counts.items():
        print(f"    {booster}: {count} failure(s)")
    
    # Count by launch site
    site_counts = df_result.groupby('LaunchSite').size()
    print(f"\n  By Launch Site:")
    for site, count in site_counts.items():
        print(f"    {site}: {count} failure(s)")
    
else:
    print("\nâŒ No drone ship failure records found for 2015")
    
    # Check what we do have for 2015
    print("\nðŸ” Checking 2015 data availability...")
    
    # Check total 2015 launches
    check_query = '''
    SELECT 
        COUNT(*) as Total2015Launches,
        SUM(CASE WHEN "LandingOutcome" LIKE '%Failure%' OR "Outcome" LIKE '%Failure%' THEN 1 ELSE 0 END) as TotalFailures,
        SUM(CASE WHEN "LandingPad" LIKE '%drone%' OR "LandingPad" LIKE '%ASDS%' THEN 1 ELSE 0 END) as DroneShipLaunches
    FROM SPACEXTBL 
    WHERE substr(Date, 0, 5) = '2015';
    '''
    
    check_result = pd.read_sql_query(check_query, con)
    
    if not check_result.empty:
        total = check_result['Total2015Launches'].iloc[0]
        failures = check_result['TotalFailures'].iloc[0]
        drone_ships = check_result['DroneShipLaunches'].iloc[0]
        
        print(f"  Total 2015 launches: {total}")
        print(f"  Total failures in 2015: {failures}")
        print(f"  Drone ship launches in 2015: {drone_ships}")

con.close()
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import pandas as pd
import sqlite3

con = sqlite3.connect('spacex.db')

# One-liner to get ranked landing outcomes
result = pd.read_sql_query(
    '''SELECT "LandingOutcome", COUNT(*) as Count, 
       RANK() OVER (ORDER BY COUNT(*) DESC) as Rank
       FROM SPACEXTBL 
       WHERE "Date" BETWEEN '2010-06-04' AND '2017-03-20' 
       AND "LandingOutcome" IS NOT NULL
       GROUP BY "LandingOutcome" 
       ORDER BY Count DESC;''', 
    con
)

print("Ranked Landing Outcomes:")
print(result.to_string(index=False))

con.close()
```
